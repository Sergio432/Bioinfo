```
import Bio
from Bio import SeqIO
import matplotlib.pyplot as plt
import numpy as np
v=SeqIO.read("https://raw.githubusercontent.com/Sergio432/Bioinfo/master/Output%20(multiFASTA).FASTA","fasta")
DNA_seqs=[repr(seq_record.seq)for seq_record in SeqIO.parse(v, "fasta")]
print(DNA_seqs)
DNA_seq_ID=[(seq_record.id) for seq_record in SeqIO.parse(v, "fasta")]
DNA_IDs=([i  for i  in  DNA_seq_ID[1:]])
DNA_Seq=([i for i in DNA_seqs[1:]])
#ids
x_axis=[str(i) for  i in  DNA_IDs]

print(x_axis)

A=[i.count("A") for i in DNA_Seq]
T=[i.count("T") for i in DNA_Seq]
G=[i.count("G") for i in DNA_Seq]
C=[i.count("C") for i in DNA_Seq]
print("A:",A)
print("T",T)
print("G",G)
print("C",C)

#gc contenido
GC_coun=0
GC_count=[]
seq_num=[]
for i in range(0, len(DNA_Seq)):
    GC_coun= round(((G[i]+C[i])/(A[i]+T[i]+G[i]+C[i]))*100 ,2)
    GC_count.append(GC_coun)
    seq_num.append(i+1)
print(GC_count)
print(len(GC_count))


plt.bar(seq_num, GC_count, color='maroon', width=0.4)

plt.xlabel('VIRUSES')
plt.ylabel("% G & C")
plt.title("% G & C  VS VIRUSES")
plt.show()
```
